```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
sns.distplot(tips['total_bill'],  rug=True, kde_kws={"shade": True}, bins= 6)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data=tips, kind="kde", color="#4ca8ed", height = 7, xlim = (0,50), ylim = (0,7))
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue ="sex", palette = "Set1")
```

```python
sns.barplot(x = "sex", y = "tip", data = tips)
```

```python
tips2 = tips.head(10)
sns.barplot(x = "sex", y = "tip", data = tips2, estimator=len)
```

```python
sns.barplot(x = "day", y = "tip", data = tips,palette="rocket", estimator=np.median)
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, palette="pastel")
```

```python
sns.boxplot(x = "day", y = "total_bill",hue="smoker", data = tips, palette="pastel")
```

```python
sns.boxplot(data = tips, palette="pastel", orient='h')
```

```python
sns.stripplot(x = "day", y = "total_bill",hue="smoker", data = tips, palette="pastel", size = 6, linewidth = 1)
```

```python
sns.stripplot(x ="day", y = "total_bill", hue ="smoker",dodge=True, data = tips, palette="pastel", alpha=.35, size = 30)
```

```python
f, ax = plt.subplots(figsize=(15, 8))
ax = sns.heatmap(tips.corr(), annot = True, linewidths=.5, cmap="YlGnBu")
```

```python
g = sns.FacetGrid(tips, col = "time", row = "smoker", hue = "day")
g = g.map(plt.hist, "total_bill")
g.add_legend(title="legend")
```
