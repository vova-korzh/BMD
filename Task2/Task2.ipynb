```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'y') #build plot
plt.xlabel('X Label') #add name of x axis
plt.ylabel('Y Label')#add name of y axis
plt.title('Title') #add title
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'y')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes.plot(x,y, 'y')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')

axes2.plot(y,x)
axes2.set_xlabel("Y label")
axes2.set_ylabel("X label")
axes2.set_title("itle2")
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'y')
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("Title 1")

axes[1].plot(y,x,'b')
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")
axes[1].set_title("Title 2")
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4 )
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4 )
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, color = 'r') #red line
ax.plot(x, x**3, color = 'black') #black line
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 2, ls = '--', marker = '*', markerfacecolor = 'red')
```
